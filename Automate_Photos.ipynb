```python
import csv
import shutil
import os
```

```python
f = open('NewDelhi/Aizawl_PM25_records_updated.csv')
f2='files_to_send/Aizawl_PM25_records_edited.csv'
#So _edited is the final version
csv_f = list(csv.reader(f))
with open(f2, 'a+', newline='') as csvFile:
    for i in range(1,len(csv_f)):
        flnm=csv_f[i][0]
        src="Aizawl/"+flnm+".jpg"
        dst="new_Aizawl"
        try:
            shutil.copy2(src, dst)
            writer = csv.writer(csvFile)
            writer.writerow(csv_f[i])
        except:
            print(flnm," image not found")

f.close()
```
